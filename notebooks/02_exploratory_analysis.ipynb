```python
# 02_exploratory_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure the save directory exists
os.makedirs('../reports/figures', exist_ok=True)

# Load data
df = pd.read_csv('../data/processed/processed_data.csv')

# Basic info
print("üîç Data Overview:")
print(df.info())
print(df.describe())

# --- 1. Churn distribution ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='churn')
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../reports/figures/churn_distribution.png')
plt.show()

# --- 2. Correlation heatmap ---
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../reports/figures/correlation_heatmap.png')
plt.show()

# --- 3. Tenure vs Churn (boxplot) ---
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='churn', y='tenure')
plt.title('Tenure by Churn')
plt.xlabel('Churn')
plt.ylabel('Tenure')
plt.tight_layout()
plt.savefig('../reports/figures/tenure_by_churn.png')
plt.show()

# --- 4. MonthlyCharges vs Churn (boxplot) ---
if 'monthlycharges' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x='churn', y='monthlycharges')
    plt.title('Monthly Charges by Churn')
    plt.xlabel('Churn')
    plt.ylabel('Monthly Charges')
    plt.tight_layout()
    plt.savefig('../reports/figures/monthly_charges_by_churn.png')
    plt.show()
```
